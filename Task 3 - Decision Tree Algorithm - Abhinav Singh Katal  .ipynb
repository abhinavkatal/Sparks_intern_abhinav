```python
# Import all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the iris datasets
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
# Storing the independent and dependent variable data
X = iris.data
y = iris.target
```

```python
# Importing the test_train_split to create a model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
# Importing Decision tree Classifier and creating an object
from sklearn.tree import DecisionTreeClassifier
dmodel = DecisionTreeClassifier()
dmodel.fit(X_train, y_train)
```

```python
# Create a variable ymodel which will store the predicted values based on the training dataset
ymodel = dmodel.predict(X_test)
ymodel
```

```python
# Importing the metrics or performance measuring classes
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test, ymodel))
```

```python
print(classification_report(y_test, ymodel))
```

```python
# Printing the model efficiency
print(accuracy_score(y_test, ymodel))
```

# Hence, our model is 96.67% accurate

```python
# Now let us give it a new data to predict
new_predict = dmodel.predict([[8.3, 6.6, 7 , 3.5]])
```

```python
new_predict
```

# When given a new reading of 8.3, 6.6, 7 , 3.5 , the model predicted that it of the type 2 i.e. Virginica type.

```python
# Create a model based on the whole data
dtree = DecisionTreeClassifier(criterion = 'gini')
dtree = dtree.fit(iris.data,iris.target)
# iris.data is the independent variable and iris.target is the dependent variable
```

```python
# Lets create a decision tree of the iris data
from sklearn import tree
plt.figure(figsize = (40,20))
tree.plot_tree(dtree, filled = True, feature_names = iris.feature_names, class_names = True, label = 'all')
```

```python
# Print the decision tree in textual form
print(tree.export_text(dtree, feature_names = iris.feature_names))
```
